```python

# Importer les bibliothèques nécessaires
import pandas as pd
import requests
import gspread
from oauth2client.service_account import ServiceAccountCredentials
import zipfile
from google.colab import files

# Télécharger le fichier config.json
print("Veuillez télécharger votre fichier config.json")
uploaded = files.upload()
config_path = list(uploaded.keys())[0]

# Télécharger le fichier ZIP
print("Veuillez télécharger votre fichier ZIP")
uploaded = files.upload()
zip_path = list(uploaded.keys())[0]

# Extraire le fichier ZIP
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('extracted_data')

extracted_file = zip_ref.namelist()[0]
csv_path = 'extracted_data/' + extracted_file

# Lire le fichier CSV extrait
dataArray = pd.read_csv(csv_path, delimiter="|")
print('Données parsées.')

# Filtrer les données
filteredData = dataArray[dataArray[5] == "10"]
print('Nombre de lignes après filtrage : ' + str(len(filteredData)))

# Ajouter les en-têtes
headers = ["identifiantRPPS", "nomExercice", "prenomExercice"]
filteredData = pd.concat([pd.DataFrame([headers]), filteredData], ignore_index=True)

# Authentification avec l'API Google Sheets
scope = ['https://spreadsheets.google.com/feeds', 'https://www.googleapis.com/auth/drive']
credentials = ServiceAccountCredentials.from_json_keyfile_name(config_path, scope)
client = gspread.authorize(credentials)

# Ouvrir la feuille de calcul et sélectionner la feuille
spreadsheetId = "1ottTPiBjgBXSZSj8eU8jYcatvQaXLF64Ppm3qOfYbbI"
sheetName = "RPPS"
spreadsheet = client.open_by_key(spreadsheetId)
sheet = spreadsheet.worksheet(sheetName)

# Effacer le contenu de la feuille
sheet.clear()

# Convertir les données filtrées en une liste de listes
data = filteredData.values.tolist()

# Insérer les données dans la feuille
sheet.insert_rows(data)

print('Données insérées dans la feuille de calcul.')
```
